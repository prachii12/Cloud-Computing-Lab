```python
# Q2

!wget https://www.gutenberg.org/cache/epub/11/pg11.txt
file = open("pg11.txt", "r", encoding="utf-8")
data = file.readlines()
```

```python
import re

count = 0

for line in data:
    words = re.findall(r'\bCheshire\b', line)
    count += len(words)

print("Cheshire count:", count)
```
