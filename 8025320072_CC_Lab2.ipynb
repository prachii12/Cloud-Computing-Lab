```python
#1. Create a Data File (20000 × 10000) Using Random Module

import numpy as np

# Define size
rows = 20000
cols = 10000

# Generate random matrix (float32 to reduce memory usage)
data = np.random.rand(rows, cols).astype(np.float32)

# Save to file
np.save("large_matrix.npy", data)

print("Data file created successfully!")
print("Shape:", data.shape)
```

```python
# 2. Assign Subsequent Rows to First Row Elements

import numpy as np

# Example matrix
matrix = np.random.randint(1, 10, (5, 5))
print("Original Matrix:\n", matrix)

# Assign first row to all rows
matrix[:] = matrix[0]

print("\nModified Matrix:\n", matrix)
```

```python
# 3. Group Similar Elements into Matrix

import numpy as np

matrix = np.random.randint(1, 20, (5, 5))
print("Original Matrix:\n", matrix)

# Group even and odd elements
even = matrix[matrix % 2 == 0]
odd = matrix[matrix % 2 != 0]

print("\nEven Elements:", even)
print("Odd Elements:", odd)
```

```python
# 4. Add and Subtract Matrices

import numpy as np

A = np.random.randint(1, 10, (3, 3))
B = np.random.randint(1, 10, (3, 3))

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)

# Addition
add = A + B
print("\nAddition:\n", add)

# Subtraction
sub = A - B
print("\nSubtraction:\n", sub)
```

```python
# 5. Create n × n Matrix Where Sum of Opposite Corners of Every Sub-matrix is Even

import numpy as np

n = 4

# Create matrix with all even numbers
matrix = np.random.randint(1, 50, (n, n)) * 2

print("Matrix:\n", matrix)
```

```python
# 6. Row-wise Element Addition in Tuple Matrix

# Tuple matrix
matrix = ((5, 2, 4),
          (4, 5, 1),
          (3, 6, 5))

print("Original Tuple Matrix:", matrix)

# Row-wise sum
row_sums = tuple(sum(row) for row in matrix)

print("Row-wise Addition:", row_sums)
```
